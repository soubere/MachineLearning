```python
import pandas as pd
data = pd.read_csv('kc_house_data.csv', sep = ',', encoding = "ISO-8859-1")
data.head()
```

```python
#Nombre totale de valeur manquante dans le data frame
data.isnull().sum().sum()
```

```python
#Afficher la relation entre le prix et l'habitation
data[["sqft_living","price"]].groupby(["sqft_living"], as_index = True).mean()
#il ya une proportionnalite positive entre le prix et l'habitation
```

```python
#le prix par rapport a la parcelle
data[["sqft_lot","price"]].groupby(["sqft_lot"], as_index = True).mean()
```

```python
#le prix par rapport a proximite de la mer waterfront and price
data[["waterfront","price"]].groupby(["waterfront"], as_index = True).mean()
```

```python
#le prix par rapport aux zipcode
data[["zipcode","price"]].groupby(["zipcode"], as_index = True).mean()
```
